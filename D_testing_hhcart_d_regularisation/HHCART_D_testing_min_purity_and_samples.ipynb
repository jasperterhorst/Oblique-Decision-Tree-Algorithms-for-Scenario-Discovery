# Min Purity and Samples Experiment

This notebook explores the effects of varying the `min_purity` and `min_samples` parameters on hierarchical hyperplane decision trees (HHCartD) applied to 2D shape datasets. Four different shape geometries (“barbell”, “radial_segment”, “rectangle”, “sine_wave”) under two noise levels (0.00 and 0.05) are investigated.

The analysis is divided into two main parts:  
1. **Minimum Purity Experiments**  
   - Runs with `min_purity = 1.00`, `0.95`, and `0.90` (keeping `min_samples = 2` and `max_depth = 8`)  
   - For each setting, HHCartD models are built and reloaded, followed by generation of:
     - Coverage vs. Density plots by tree depth
     - Coverage vs. Density plots colored by Class 1 leaf count
     - Oblique split overlays at depths 4, 6, and 8  

2. **Minimum Samples Experiments**  
   - Runs with `min_samples = 0.01` and `0.05` (keeping `min_purity = 1.00` and `max_depth = 8`)  
   - For each setting, HHCartD models are built and reloaded, followed by generation of the same set of plots as above  

Each experiment section loads the shape datasets, trains the HHCartD model, and saves the resulting PDF plots for later inspection. This structure facilitates a systematic comparison of how purity and sample‐size constraints influence tree structure and classification performance.  


## Imports

```python
from itertools import product

from HHCART_SD import HHCartD, load_model
from src.load_shapes import load_shape_dataset
```

## Experiment (min_purity = 1.0, mass_min = 2)

```python
# Configuration
shapes = ["barbell", "radial_segment", "rectangle", "sine_wave", "star"]

noise_levels = ["000", "005"]
dataset_names = [f"{shape}_2d_label_noise_{noise}" for shape, noise in product(shapes, noise_levels)]
folder_name = "shapes"
max_depth = 8
min_purity = 1.0
mass_min = 2
purity_str = str(min_purity).replace('.', '_')
samples_str = str(mass_min).replace('.', '_')

# Load datasets
datasets = load_shape_dataset(folder_name=folder_name)

# Helpers
def format_shape_and_noise(dataset_name: str) -> str:
    parts = dataset_name.split("_")
    shape = parts[0].capitalize().replace("segment", "Segment").replace("wave", "Wave")
    noise = parts[-1]
    noise_float = float(noise) / 100
    return f"{shape} with Boundary Noise {noise_float:.2f}"

# Main loop
for dataset_name in dataset_names:
    if dataset_name not in datasets:
        continue

    X, y = datasets[dataset_name]
    descriptor = format_shape_and_noise(dataset_name)
    save_id = f"{dataset_name}_depth_{max_depth}_p_{purity_str}_s_{samples_str}"

    hh = HHCartD(X, y, min_purity=min_purity, mass_min=mass_min, max_depth=max_depth)
    hh.build_tree(save_id)
    hh = load_model(save_id)
    
    # Shorten only the filename key for radial_segment
    if dataset_name.startswith("radial_segment"):
        fname_key = dataset_name.replace("radial_segment", "radialseg")
    else:
        fname_key = dataset_name

    # Plot 1: Coverage vs Density by Depth
    f1 = f"cvd_depth_{fname_key}_d{max_depth}_p{min_purity}_s{mass_min}.pdf"
    t1 = f"Coverage vs Density by Tree Depth\n({descriptor})"
    hh.plot_tradeoff_path(save=True, filename=f1, title=t1)

    # Plot 2: Coverage–Density vs Class 1 Leaf Count
    f2 = f"cvd_cls1_{fname_key}_d{max_depth}_p{min_purity}_s{mass_min}.pdf"
    t2 = f"Coverage–Density vs Class 1 Leaf Count\n({descriptor})"
    hh.plot_tradeoff_path(save=True, color_by="class1_leaf_count", filename=f2, title=t2)

    # Plot 3: Oblique Splits at multiple depths
    actual_max = max(hh.available_depths())
    for d in [4, 6, 8]:
        depth_to_plot = min(d, actual_max)
        f_overlay = f"split_{fname_key}_d{depth_to_plot}_p{min_purity}_s{mass_min}.pdf"
        t_overlay = f"Oblique Splits at Depth {depth_to_plot}\n({descriptor})"
        hh.plot_splits_2d_overlay(
            depth=depth_to_plot,
            save=True,
            filename=f_overlay,
            title=t_overlay
        )

    # Plot 4: Metrics vs Tree Depth
    f3 = f"metrics_depth_{fname_key}_d{max_depth}_p{min_purity}_s{mass_min}.pdf"
    t3 = f"Performance Metrics vs. Tree Depth\n({descriptor})"
    hh.plot_metrics_vs_structure(save=True, filename=f3, title=t3, x_axis="depth", y_axis=["coverage", "density"])

    # Plot 5: Metrics vs Class 1 Leaf Count
    f4 = f"metrics_cls1_{fname_key}_d{max_depth}_p{min_purity}_s{mass_min}.pdf"
    t4 = f"Performance Metrics vs. Class 1 Leaf Count\n({descriptor})"
    hh.plot_metrics_vs_structure(save=True, filename=f4, title=t4, x_axis="class1_leaf_count", y_axis=["coverage", "density"])
```
